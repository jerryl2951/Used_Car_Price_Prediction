### Outliers Treatment

```python
import seaborn as sns
import matplotlib.pyplot as plt
for i in data.drop(['selling_price','seats'],axis=1).select_dtypes(include=['int64','float64']).columns:
    sns.boxplot(data[i])
    plt.title(i)
    plt.show()
```

```python
for i in data.drop(['selling_price','seats'],axis=1).select_dtypes(include=['int64','float64']).columns:
    print(i)
    desc=data[i].describe()
    iqr=desc['75%']-desc['25%']
    ul=desc['75%']+1.5*iqr
    ll=desc['25%']-1.5*iqr
    data[i][data[i]>ul]=ul
    data[i][data[i]<ll]=ll
```

### Importing packages for performing machine learning algorithms on datasets

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25) #splitting the datasets as training datasets and testing dataset
```

```python
regressor=LinearRegression() #declaring an object of LinearRegression algorithm
```

```python
regressor.fit(X_train,y_train) #fitting the training datasets to the linear regression object
```

```python
ypred=regressor.predict(X_test)  # predicting the values using linear Regression
ypred #predicted values
```

```python
y_test   #actual values
```

```python
regressor.predict(X_train)
```

```python
regressor.score(X_test,y_test) #computes the score of  Linear regresion ML model (testing accuracy)
```

```python
regressor.score(X_train,y_train) #(training accuracy)
```

```python
import math
from sklearn.metrics import mean_squared_error
math.sqrt(mean_squared_error(ypred,y_test))  #root mean square error
```
